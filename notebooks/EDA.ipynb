# Análise Exploratória de Dados — Anúncios de Venda de Veículos (EUA)

Este notebook faz uma exploração inicial do conjunto de dados `vehicles_us.csv`.

```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv')
car_data.head()
```

```python
car_data.info()
```

## Histograma da quilometragem (odometer)

```python
fig = px.histogram(car_data, x='odometer')
fig.show()
```

## Relação entre quilometragem (odometer) e preço (price)

```python
fig = px.scatter(car_data, x='odometer', y='price')
fig.show()
```
