# Análise Exploratória de Dados — Anúncios de Venda de Veículos (EUA)

Este notebook faz uma exploração inicial do conjunto de dados `vehicles_us.csv`, analisando distribuições de preço e quilometragem, e a relação entre elas.

## 1. Carregamento e inspeção inicial dos dados

```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles_us.csv')
print('Shape do dataset:', car_data.shape)
print('\nPrimeiras linhas:')
car_data.head()
```

```python
print('Informações sobre o dataset:')
car_data.info()
print('\nEstatísticas descritivas:')
car_data.describe()
```

## 2. Análise da distribuição de preços

Vamos examinar como os preços estão distribuídos no dataset.

```python
print('Estatísticas de preço:')
print(f'Preço mínimo: ${car_data["price"].min():,.0f}')
print(f'Preço máximo: ${car_data["price"].max():,.0f}')
print(f'Preço médio: ${car_data["price"].mean():,.0f}')
print(f'Preço mediano: ${car_data["price"].median():,.0f}')

fig = px.histogram(car_data, x='price', nbins=50, title='Distribuição de Preços dos Veículos')
fig.update_xaxes(title_text='Preço (USD)')
fig.update_yaxes(title_text='Frequência')
fig.show()
```

**Observação:** A maioria dos veículos está concentrada em um intervalo de preço mais baixo, com alguns outliers em valores muito altos. A distribuição é assimétrica, com cauda longa à direita.

## 3. Análise da distribuição de quilometragem (odometer)

Agora vamos ver como a quilometragem dos veículos está distribuída.

```python
print('Estatísticas de quilometragem:')
print(f'Quilometragem mínima: {car_data["odometer"].min():,.0f} km')
print(f'Quilometragem máxima: {car_data["odometer"].max():,.0f} km')
print(f'Quilometragem média: {car_data["odometer"].mean():,.0f} km')
print(f'Quilometragem mediana: {car_data["odometer"].median():,.0f} km')

fig = px.histogram(car_data, x='odometer', nbins=50, title='Distribuição de Quilometragem dos Veículos')
fig.update_xaxes(title_text='Quilometragem (km)')
fig.update_yaxes(title_text='Frequência')
fig.show()
```

**Observação:** A quilometragem também mostra uma distribuição bastante espalhada, com veículos desde 0 km até 300 mil km. Não há um pico concentrado em um único intervalo.

## 4. Relação entre quilometragem (odometer) e preço

Vamos analisar se existe uma correlação entre quilometragem e preço — veículos com mais quilometragem tendem a ser mais baratos?

```python
# Calcular correlação
correlacao = car_data['odometer'].corr(car_data['price'])
print(f'Correlação entre odometer e price: {correlacao:.3f}')

fig = px.scatter(car_data, x='odometer', y='price', title='Relação entre Quilometragem e Preço')
fig.update_xaxes(title_text='Quilometragem (km)')
fig.update_yaxes(title_text='Preço (USD)')
fig.show()
```

**Observação:** Existe uma correlação negativa entre quilometragem e preço (quanto maior a quilometragem, menor o preço, em geral). Isso faz sentido, pois veículos com mais uso tendem a valer menos. No entanto, há bastante dispersão nos dados — a relação não é perfeitamente linear.

## 5. Análise por modelo de veículo

Vamos ver quais são os modelos mais comuns no dataset.

```python
print('Top 10 modelos mais anunciados:')
print(car_data['model'].value_counts().head(10))

fig = px.bar(car_data['model'].value_counts().head(10).reset_index(), 
             x='count', y='model', orientation='h',
             title='Top 10 Modelos de Veículos mais Anunciados')
fig.update_xaxes(title_text='Quantidade')
fig.update_yaxes(title_text='Modelo')
fig.show()
```

**Observação:** Alguns modelos aparecem muito mais vezes que outros no dataset, indicando maior popularidade no mercado de carros usados.

## Conclusões

1. **Preços**: A maioria dos veículos tem preço entre 5k e 20k USD, com alguns outliers caros.
2. **Quilometragem**: Há uma grande variação de quilometragem, desde veículos quase novos até muito rodados.
3. **Correlação**: Veículos com maior quilometragem tendem a ter preços mais baixos, confirmando uma relação negativa.
4. **Modelos**: Alguns modelos são muito mais comuns nos anúncios que outros.

Estes insights informam as visualizações criadas no dashboard interativo (histogramas e scatter plots) que permitem explorar esses dados de forma dinâmica.
